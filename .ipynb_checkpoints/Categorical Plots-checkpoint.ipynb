```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### Count/Frequency Plot

```python
sns.countplot(x="sex",data=tips,palette="coolwarm")
```

### Bar Plot

```python
sns.barplot(x="smoker", y="tip", data=tips, estimator = np.median)
```

### Box Plots

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

### Violin Plots

```python
sns.violinplot(x="day", y="tip", data=tips, hue="smoker", split=True)
```

### Strip Plots

```python
sns.stripplot(x="day", y="tip", data=tips, hue="smoker")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="smoker", jitter = True, split=True)
```

### Swarm Plots

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

### Combine strip and box plot

```python
sns.stripplot(x="day", y="tip", data=tips, color="black")
sns.boxplot(x="day", y="tip", data=tips, boxprops=dict(alpha=.3))
```
